Code for opening H5 files and storing the data in a CSV file

```python
import os
import tables
import pandas as pd
```

```python
path = os.getcwd()+"/data/MillionSongSubset"
ct = 0
storage_df = pd.DataFrame()
for path_fwd in os.listdir(path):
    for path_fwd1 in os.listdir(path+"/"+path_fwd):
        for path_fwd2 in os.listdir(path+"/"+path_fwd+"/"+path_fwd1):
            for path_fwd3 in os.listdir(path+"/"+path_fwd+"/"+path_fwd1+"/"+path_fwd2):
                print(path_fwd3)
                h5 = tables.open_file(path+"/"+path_fwd+"/"+path_fwd1+"/"+path_fwd2+"/"+path_fwd3, mode='r')
                ct+=1
                store_dict = {}
                store_dict['INDEX'] = ct
                store_dict['YEAR'] = h5.root.musicbrainz.songs.cols.year[0]
                store_dict['TEMPO'] = h5.root.analysis.songs.cols.tempo[0]
                store_dict['START_OF_FADE_OUT'] = h5.root.analysis.songs.cols.start_of_fade_out[0]
                store_dict['LOUDNESS'] = h5.root.analysis.songs.cols.loudness[0]
                store_dict['END_OF_FADE_IN'] = h5.root.analysis.songs.cols.end_of_fade_in[0]
                store_dict['DURATION'] = h5.root.analysis.songs.cols.duration[0]
                store_dict['AUDIO_MD5'] = h5.root.analysis.songs.cols.audio_md5[0]
                
                store_dict['NUM_SONGS'] = h5.root.metadata.songs.nrows
                store_dict['ARTIST_FAMILIARITY'] = h5.root.metadata.songs.cols.artist_familiarity[0]
                store_dict['ARTIST_HOTTTNESSS'] = h5.root.metadata.songs.cols.artist_hotttnesss[0]
                store_dict['ARTIST_ID'] = h5.root.metadata.songs.cols.artist_id[0]
                store_dict['ARTIST_MBID'] = h5.root.metadata.songs.cols.artist_mbid[0]
                store_dict['ARTIST_PLAYMEID'] = h5.root.metadata.songs.cols.artist_playmeid[0]
                store_dict['ARTIST_7DIGITALID'] = h5.root.metadata.songs.cols.artist_7digitalid[0]
                store_dict['ARTIST_LATITUDE'] = h5.root.metadata.songs.cols.artist_latitude[0]
                store_dict['ARTIST_LONGITUDE'] = h5.root.metadata.songs.cols.artist_longitude[0]
                store_dict['ARTIST_LOCATION'] = h5.root.metadata.songs.cols.artist_location[0]
                store_dict['ARTIST_NAME'] = h5.root.metadata.songs.cols.artist_name[0]
                store_dict['RELEASE'] = h5.root.metadata.songs.cols.release[0]
                store_dict['RELEASE_7DIGITALID'] = h5.root.metadata.songs.cols.release_7digitalid[0]
                store_dict['SONG_ID'] = h5.root.metadata.songs.cols.song_id[0]
                store_dict['SONG_HOTTTNESSS'] = h5.root.metadata.songs.cols.song_hotttnesss[0]
                store_dict['TITLE'] = h5.root.metadata.songs.cols.title[0]
                store_dict['TRACK_7DIGITALID'] = h5.root.metadata.songs.cols.track_7digitalid[0]
                if h5.root.metadata.songs.nrows == 1:
                    store_dict['SIMILAR_ARTISTS'] = h5.root.metadata.similar_artists[h5.root.metadata.songs.cols.idx_similar_artists[0]:]
                    store_dict['SEGMENTS_START'] = h5.root.analysis.segments_start[h5.root.analysis.songs.cols.idx_segments_start[0]:]
                    store_dict['SEGMENT_PITCHES'] = h5.root.analysis.segments_pitches[h5.root.analysis.songs.cols.idx_segments_pitches[0]:,:]
                    store_dict['SEGMENT_TIMBRE'] = h5.root.analysis.segments_timbre[h5.root.analysis.songs.cols.idx_segments_timbre[0]:,:]
                    store_dict['SEGMENT_LOUDNESS_MAX'] = h5.root.analysis.segments_loudness_max[h5.root.analysis.songs.cols.idx_segments_loudness_max[0]:]
                    store_dict['BEATS_START'] = h5.root.analysis.beats_start[h5.root.analysis.songs.cols.idx_beats_start[0]:]
                    store_dict['BARS_START'] = h5.root.analysis.bars_start[h5.root.analysis.songs.cols.idx_bars_start[0]:]
                    store_dict['TATUMS_START'] = h5.root.analysis.tatums_start[h5.root.analysis.songs.cols.idx_tatums_start[0]:]
                    
                else:
                    store_dict['SIMILAR_ARTISTS'] = h5.root.metadata.similar_artists[h5.root.metadata.songs.cols.idx_similar_artists[0]:
                                            h5.root.metadata.songs.cols.idx_similar_artists[1]]
                    store_dict['SEGMENTS_START'] = h5.root.analysis.segments_start[h5.root.analysis.songs.cols.idx_segments_start[0]:
                                           h5.root.analysis.songs.cols.idx_segments_start[1]]
                    store_dict['SEGMENT_PITCHES'] = h5.root.analysis.segments_pitches[h5.root.analysis.songs.cols.idx_segments_pitches[0]:
                                             h5.root.analysis.songs.cols.idx_segments_pitches[1],:]
                    store_dict['SEGMENT_TIMBRE'] = h5.root.analysis.segments_timbre[h5.root.analysis.songs.cols.idx_segments_timbre[0]:
                                            h5.root.analysis.songs.cols.idx_segments_timbre[1],:]
                    store_dict['SEGMENT_LOUDNESS_MAX'] = h5.root.analysis.segments_loudness_max[h5.root.analysis.songs.cols.idx_segments_loudness_max[0]:
                                                  h5.root.analysis.songs.cols.idx_segments_loudness_max[1]]
                    store_dict['BEATS_START'] = h5.root.analysis.beats_start[h5.root.analysis.songs.cols.idx_beats_start[0]:
                                        h5.root.analysis.songs.cols.idx_beats_start[1]]
                    store_dict['BARS_START'] = h5.root.analysis.bars_start[h5.root.analysis.songs.cols.idx_bars_start[0]:
                                       h5.root.analysis.songs.cols.idx_bars_start[1]]
                    store_dict['TATUMS_START'] = h5.root.analysis.tatums_start[h5.root.analysis.songs.cols.idx_tatums_start[0]:
                                         h5.root.analysis.songs.cols.idx_tatums_start[1]]
                    
                    
                
                storage_df = storage_df.append(store_dict,ignore_index=True)
                if ct%500 == 0:
                    storage_df.to_csv("SOURCE-DATASET-1.csv",index=False)
                    
storage_df.to_csv("SOURCE-DATASET-1.csv",index=False)
```
